```python
import pandas as pd
star_wars = pd.read_csv('star_wars.csv',encoding='ISO-8859-1')
```

```python
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars.head(10)
```

```python
yes_no = {"Yes": True, "No": False}
col =[
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]
for c in col:
    star_wars[c] = star_wars[c].map(yes_no)

star_wars.head()
```

```python
import numpy as np
value_conversion = {"Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
                   }

for c in star_wars.columns[3:9]:
    star_wars[c] = star_wars[c].map(value_conversion)
   
                   
```

```python
star_wars = star_wars.rename(columns={'Which of the following Star Wars films have you seen? Please select all that apply.':'seen_1', 
                    'Unnamed: 4':'seen_2',
                    'Unnamed: 5':'seen_3',
                    'Unnamed: 6':'seen_4',
                    'Unnamed: 7':'seen_5',
                    'Unnamed: 8':'seen_6'})
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'ranking_1',
                                     'Unnamed: 10':'ranking_2',
                                     'Unnamed: 11': 'ranking_3',
                                     'Unnamed: 12': 'ranking_4',
                                     'Unnamed: 13': 'ranking_5',
                                      'Unnamed: 14': 'ranking_6'
                            })
```

```python
star_wars.iloc[:,9:15]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.bar(range(6),star_wars[star_wars.columns[9:15]].mean())
```

```python
#as we can see this bar graph illustrates the average ranking per 6 movies in the star wars series, the fifth movie seems to have the highest ranking.
```

```python
plt.bar(range(6),star_wars[star_wars.columns[3:9]].sum())
```

```python
#our data shows a correlation between rating and number of people who saw the movie, as we see the 4th movie had the highest rating and also the highest number of people who saw the movie.
```

```python
males = star_wars[star_wars['Gender']== 'Male']
females = star_wars[star_wars['Gender']== 'Female']
```

```python
plt.bar(range(6),males[males.columns[9:15]].mean())
plt.show()
plt.bar(range(6),females[females.columns[9:15]].mean())
plt.show()
```

```python
plt.bar(range(6), males[males.columns[3:9]].sum())
plt.show()

plt.bar(range(6), females[females.columns[3:9]].sum())
plt.show()
```

```python
#the male ratings are a bit odd because they saw movies 1-3 the most but did not like them too much
```

```python
#the sweet spot for females were the 2nd and 3rd movies with a declining preference from there on
```

```python
star_wars['Location (Census Region)'].unique()
```

```python
import numpy
star_wars = star_wars.groupby('Location (Census Region)').agg(numpy.mean)
star_wars
```

```python
rank_cols = ['ranking_1','ranking_2','ranking_3','ranking_4','ranking_5','ranking_6']
cols = ['East North Central','East South Central','Middle Atlantic','Mountain','New England','Pacific','South Atlantic','West North Central','West South Central']
movies = ['Star Wars: Episode I The Phantom Menace','Star Wars: Episode II Attack of the Clones','Star Wars: Episode III Revenge of the Sith','Star Wars: Episode IV A New Hope','Star Wars: Episode V The Empire Strikes Back','Star Wars: Episode VI Return of the Jedi']
tick_positions = range(0,9)
for c in rank_cols:
    fig,ax=plt.subplots()
    ax.bar(range(9),star_wars[c])
    ax.set_xticks(tick_positions)
    ax.set_xticklabels(cols,rotation=45)
    ax.set_xlabel('Region')
    ax.set_ylabel('Rating')
 
plt.show()
```

```python
#grouped the data by region and found the means of all numerical data values
#created bar graphs for each of the 6 Star Wars movies based on their respective ratings
#to reiterate a rating of 1 is the highest with 6 being the lowest
#from the first star wars movie west south central liked the most
#second and third star wars movies were most liked by pacific
#fourth and fifth star wars movies were liked most by East South Central
#sixth star wars movie was liked by New England the most
```
